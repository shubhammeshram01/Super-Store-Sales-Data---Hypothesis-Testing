# Hypothesis Testing 

## 7) Hypothesis 1 : Technology products have the highest profit margins as compared to the other product categories

```python
data.head(2)
```

```python
data.groupby("Category")["Profit"].sum()
```

```python
cat_profit = data.groupby("Category")["Profit"].sum()
py.figure(figsize = (12 , 5))
cat_profit.plot(kind = "bar", edgecolor = "black")
py.xlabel("Category")
py.ylabel("Profit")
py.title("Profit by Categories")
py.xticks(rotation = "0")
py.show()

# Conclusion : The hypothesis is supported as technology products have highest profit margins as compared to other categories.
```

## 8) Hypothesis 2 : The east region has the highest sales compared to the other regions

```python
data.head(2)
```

```python
py.figure(figsize = (12,5))
data.groupby("Region")["Sales"].sum().plot(kind = "bar")
py.title("Sales by region")
py.ylabel("Sales")
py.show()

# Conclusion : The hypothesis is not supported as Central region has the highest sales not the East region
```

## 10) Hypothesis 4 : Orders with same-day shipping have the lowest rate of returned products

```python
data.head(2)
```

```python
data.groupby("Ship Mode").size()
```

```python
data["Ship Mode"].value_counts()
```

### Check total order by shipping mode

```python
total_orders_by_shipping_mode = data.groupby("Ship Mode").size()
```

```python
data[data['Profit'] < 0 ].groupby("Ship Mode").size()
```

### Check for returned orders shipping mode

```python
returned_orders_by_shipping_mode = data[data["Profit"] < 0].groupby("Ship Mode").size()
```

```python
returned_orders_by_shipping_mode
```

### Returned percentage by shipping mode 

```python
(returned_orders_by_shipping_mode / total_orders_by_shipping_mode) * 100
```

```python
returned_percentage_per_shipping_mode = (returned_orders_by_shipping_mode / total_orders_by_shipping_mode) * 100
```

```python
returned_percentage_per_shipping_mode
```

```python
py.figure(figsize = (12 , 5))
returned_percentage_per_shipping_mode.plot(kind = "bar" , color = "lightgreen" , edgecolor = "black")
py.xticks(rotation = "0")
py.title("Return percentage by shipping mode")
py.show()

# Conclusion : The hypothesis is supported as same day has lowest rate of returned products 
```

## 11) Hypothesis 5 : The company's profit is more on weekdays than on weekends

```python
data.head(2)
```

```python
pd.DatetimeIndex(data["Order Date"]).day_name()
```

```python
data["Order Day"] = pd.DatetimeIndex(data["Order Date"]).day_name()
```

```python
data.head(2)
```

```python
data.groupby("Order Day")["Profit"].sum()
```

```python
day_profit = data.groupby("Order Day")["Profit"].sum()

day_profit
```

```python
py.figure(figsize=(12, 5))
day_profit.plot(kind = "bar" , color = "brown", edgecolor = "black")
py.title("Total profit by day")
py.xlabel("Day of the week")
py.ylabel("Total profit")
py.xticks(rotation = "0")
py.show()


# Conclusion : The hypothesis is supported as compnay's profit is more on weekdays than weekends
```
